# Customer Behaviour, Risk and Decision Analytics

## Objective

This project analyses a bank customer dataset to identify patterns linked to customer churn. The aim is to understand which customer behaviours and characteristics may be associated with customers leaving the bank.

The project focuses on using data analysis to support decision-making, while also considering data quality, limitations, and responsible interpretation of results.



```python
import pandas as pd

df = pd.read_csv("/var/local/Bank Customer Churn Prediction.csv")
df.head()
```

## Data Overview

This section explores the structure of the dataset, including the number of rows, columns, column names, and data types.

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

The dataset contains customer-level information, where each row represents one bank customer.
The target variable is "churn", which indicates whether a customer left the bank. A value of 1 means the customer churned, while 0 means the customer stayed.

## Data Quality Check

Before analysing the data, I checked whether there were any missing values.

```python
df.isnull().sum()
```

The dataset contains no missing values, as all columns returned 0 when checked.

This means no missing data needed to be removed or filled in. However, the absence of missing values does not guarantee that the dataset is complete in a wider business context. Important factors such as customer satisfaction, service quality or external financial pressures may not be included.

## Churn Distribution

This section looks at how many customers stayed compared to how many churned.

```python
df["churn"].value_counts()
```

```python
df["churn"].value_counts(normalize=True)*100
```

The dataset shows that 7,963 customers stayed, while 2,037 customers churned.

This means approximately 20% of customers left the bank. Understanding the churn rate is important because it provides a baseline for analysing customer retention.

## Age and Churn

This section compares the average age of customers who stayed with customers who churned.

```python
df.groupby("churn")["age"].mean()
```

Customers who stayed have an average age of 37, while customers who churned have an average age of 45.

This suggests that older customers may be more likely to churn. However, this does not prove that age causes churn. It shows a pattern that may be worth investigating further.

##Tenure and Churn

This section compares how long customers have been with the bank and whether this appears linked to churn.

```python
df.groupby("churn")["tenure"].mean()
```

Customers who stayed have an average tenure of 5.03 years, while customers who churned have an average tenure of 4.93 years.

The difference is very small, suggesting that tenure does not appear to be a strong factor influencing churn in this dataset.

##Activity and Churn

This section explores whether active customers are less likely to churn.

```python
df.groupby("churn")["active_member"].mean()
```

Approximately 55% of customers who stayed were active members, compared with approximately 36% of customers who churned.

This suggests that customer acitivty is strongly associated with churn. Less active customers appear more likely to leave, which could indicate that customer engagement is an improtant factor in retention.

## Account Balance and Churn

This section compares the average account balance of customers who stayed and customers who churned.

```python
df.groupby("churn")["balance"].mean()
```

Customers who stayed had an average balance of approximately 72,745, while customers who churned had a higher average balance of approximately 91,108.

This suggests that customers with higher balances may be more likely to churn. This could be important for the bank because losing higher-balance customers may have a greater financial impact.

## Summary of Findings

The main findings from the analysis are:

- Around 20% of customers in the dataset churned.
- Customers who churned were older on average than customers who stayed.
- Tenure was very similar between both groups, suggesting it may not be a strong churn factor.
- Customers who churned were less likely to be active members.
- Customers who churned had a higher average balance than those who stayed.

The strongest patterns appear to be linked to customer activity and account balance.

## Data Reliability and Risk Considerations

Although the dataset contains no missing values, it may not include all factors that influence customer behaviour.

For example, the dataset does not appear to include customer satisfaction, complaints, service quality, product usage history, or external financial pressures. These missing contextual factors may limit how fully churn can be understood.

Relying only on the available data could lead to oversimplified conclusions. In a real banking environment, further analysis using a wider variety of data would be needed before making decisions about customer retention.

This is especially important because decisions based on customer data should be fair, explainable, and responsible.

## Conclusion

This project analysed a bank customer dataset to identify patterns associated with customer churn.

The analysis found that approximately 20% of customers churned.
